```python
from __future__ import print_function
import os
import numpy as np
import pandas as pd

# file 
data_path = ['data']
filepath = os.sep.join(data_path + ['UserA.txt'])

# columns
names = ['Start Date', 'Start Time', 'End Date', 'End Time', 'Activity']

# read data
data = pd.read_csv(filepath, sep='\s+', names=names, skiprows=2)
data['Start_Time'] = data['Start Date'] + ' ' + data['Start Time']
data['End_Time'] = data['End Date'] + ' ' + data['End Time']

# clean columns
data.drop(['Start Date', 'Start Time', 'End Date', 'End Time'], axis=1, inplace=True)

# convert datatime columns
data['Start_Time'] = pd.to_datetime(data['Start_Time'])
data['End_Time'] = pd.to_datetime(data['End_Time'])
data.head(5)
```

```python
d = data.copy(deep=True)
d.head(5)
```

```python
d['Day_of_Week'] = d['Start_Time'].map(lambda x: x.dayofweek)
d['Is_Weekend'] = d['Day_of_Week'].map(lambda x: 1 if x in [5, 6] else 0)
#d['Duration_H'] = pd.to_timedelta(d['End_Time']) - pd.to_timedelta(d['Start_Time'])
d['Duration_S'] = (d['End_Time'] - d['Start_Time']).map(lambda x: x.seconds)

d.head(5)

d.groupby('Activity').Duration_S.describe()
```

```python
d.groupby('Activity').describe()
```

```python

# Filter
sd = d[d['Activity']=='Sleeping']
dd=pd.DataFrame(sd.values,columns=['Activity', 'Start', 'End', 'D', 'U', 'Duration'])
dd.drop(['D', 'U'], axis=1, inplace=True)
dd.head()
```

```python
dd['Hour'] = dd['Start'].map(lambda x: x.hour + (x.minute/60))
dd['Length'] = dd['Duration'].map(lambda x: x/60)
dd.head()
```

```python
import matplotlib.pyplot as plt

plt.scatter(dd['Hour'], dd['Length'])
plt.xlim(0, 12)
plt.xticks(range(12))
plt.ylim(240, 720)
plt.show()
```
